```python
first_names = ["ben", "roy", "julia", "larissa"]
last_names = ["clark", "brown", "johnson", "miletti"]

from random import choice

lots_of_names = []
for i in range(1000):
  lots_of_names.append([choice(first_names) + " " + choice(last_names), 
                        choice(first_names) + " " + choice(last_names)])

df = pd.DataFrame(lots_of_names, columns=["person1", "person2"])

df
```

```python
heart = pd.read_csv("https://raw.githubusercontent.com/jayachandru001/Heart-Failure-Prediction-/main/heart.csv")
heart
```

```python
heart["Sex"] = heart["Sex"].apply(lambda x: 1 if x == "M" else 0)
```

```python
heart
```

```python
heart["Sex"].mean()
```

```python
heart[heart["HeartDisease"] == 0]["Age"].hist()
```

```python
heart[heart["HeartDisease"] == 1]["Age"].hist()
```

```python
heart
```

```python
grouper = heart.groupby("HeartDisease")

grouper.mean()
```

```python
grouper = heart.groupby("ChestPainType")

grouper.mean()
```

```python
bins = [0, 20, 30, 40, 50, 60, 70, 100]
labels = ["<20","20-29","30-39","40-49","50-59","60-69","70+"]
heart["AgeGroup"] = pd.cut(heart["Age"], bins=bins, labels=labels, right=False)

heart.groupby("AgeGroup").mean()
```
